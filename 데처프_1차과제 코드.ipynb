(2) 수원시 거주 20대가 카드를 이용한 수 / 그 외

```python
index1=bc_card[(bc_card['CSTMR_CTY_RGN_NM'] != '수원시')&(bc_card['AGE_VAL'] != '20대')].index
suwon_bc_card=bc_card.drop(index1)
suwon_bc_card.reset_index(drop=True, inplace=True)
len(suwon_bc_card)
```

```python
index2=bc_card[(bc_card['CSTMR_CTY_RGN_NM'] == '수원시')&(bc_card['AGE_VAL'] == '20대')].index
nonsuwon_bc_card=bc_card.drop(index2)
nonsuwon_bc_card.reset_index(drop=True, inplace=True)
len(nonsuwon_bc_card)
```

(3) 수원시 거주 20대 중 업종별 이용금액과 이용건수 / 그 외



```python
suwon_shop = pd.pivot_table(suwon_bc_card, index=['TP_BUZ_NM'], aggfunc='sum')
suwon_shop = suwon_shop.drop(['ADMI_CTY_NO','CTY_RGN_NO','TP_GRP_NO','MAIN_BUZ_CODE','REG_YYMM','CSTMR_CTY_RGN_NO','CSTMR_MEGA_CTY_NO','FLC','MEGA_CTY_NO','TP_BUZ_NO','SEX_CTGO_CD'], axis=1)
suwon_shop = suwon_shop.sort_values(by=['AMT','CNT'], ascending=[False, False])
suwon_shop.head()
```

```python
nonsuwon_shop = pd.pivot_table(nonsuwon_bc_card, index=['TP_BUZ_NM'], aggfunc='sum')
nonsuwon_shop = nonsuwon_shop.drop(['ADMI_CTY_NO','CTY_RGN_NO','TP_GRP_NO','MAIN_BUZ_CODE','REG_YYMM','CSTMR_CTY_RGN_NO','CSTMR_MEGA_CTY_NO','FLC','MEGA_CTY_NO','TP_BUZ_NO','SEX_CTGO_CD'], axis=1)
nonsuwon_shop = nonsuwon_shop.sort_values(by=['AMT','CNT'], ascending=[False, False])
nonsuwon_shop.head()
```

(4) 영등포에서의 한식 이용금액대 서양음식 이용금액

(5) 서울시민 대 경기도민 이용건수

```python
index5 = bc_card[bc_card['CSTMR_MEGA_CTY_NM']!='서울특별시'].index
bc_card_seoul = bc_card.drop(index5)
bc_card_seoul.reset_index(drop=True, inplace=True)
index6 = bc_card[bc_card['CSTMR_MEGA_CTY_NM']!='경기도'].index
bc_card_gyeong = bc_card.drop(index6)
bc_card_gyeong.reset_index(drop=True, inplace=True)
print(bc_card_seoul['CNT'].sum(axis=0)) #서울 이용건수
print(bc_card_gyeong['CNT'].sum(axis=0)) #경기도 이용건수
```

(9) 나이대별 최대 이용 업종

```python
agebuz = pd.pivot_table(age_bc_card, index = ['AGE_VAL','TP_BUZ_NM'], aggfunc = 'sum')
agebuz = agebuz.drop(['ADMI_CTY_NO','CTY_RGN_NO','TP_GRP_NO','MAIN_BUZ_CODE','REG_YYMM','CSTMR_CTY_RGN_NO','CSTMR_MEGA_CTY_NO','FLC', 'MEGA_CTY_NO','TP_BUZ_NO','SEX_CTGO_CD'], axis=1)
agebuz = agebuz.reset_index()
agebuz = agebuz.sort_values(by=['AGE_VAL','AMT', 'CNT'], ascending=[True, False, False])
idx = agebuz.groupby('AGE_VAL')['AMT'].idxmax()
agebuz_max_spending = agebuz.loc[idx]
print(agebuz_max_spending)
```

```python
agebuz2 = pd.pivot_table(age_bc_card2, index = ['AGE_VAL','TP_BUZ_NM'], aggfunc = 'sum')
agebuz2 = agebuz2.drop(['ADMI_CTY_NO','CTY_RGN_NO','TP_GRP_NO','MAIN_BUZ_CODE','REG_YYMM','CSTMR_CTY_RGN_NO','CSTMR_MEGA_CTY_NO','FLC', 'MEGA_CTY_NO','TP_BUZ_NO','SEX_CTGO_CD'], axis=1)
agebuz2 = agebuz2.reset_index()
agebuz2 = agebuz2.sort_values(by=['AGE_VAL','AMT', 'CNT'], ascending=[True, False, False])
idx2 = agebuz2.groupby('AGE_VAL')['AMT'].idxmax()
agebuz_max_spending2 = agebuz2.loc[idx2]
print(agebuz_max_spending2)
```
